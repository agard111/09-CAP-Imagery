```python
import pandas as pd
import geopandas as gpd
from matplotlib import pyplot as plt
import contextily as ctx
```

```python
fname = "likely_track_update.geojson"
df = gpd.read_file(fname)
```

```python
storm = df.to_crs(epsg=3857).plot(figsize=(10,10), alpha = 0.5)
ctx.add_basemap(storm)
```
